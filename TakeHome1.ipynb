#### Part D) - RFA_2F and RFA_2A are marketing codes summarizing the recency and frequency of historical customer response. What values do they take and what is the frequency of each value in the data?

```python
import json


data = open("assess1_data.json", "r")

RFA_2F = dict()
RFA_2A = dict()
WEALTH_INDEX = []

count = 0

for line in data:
    info = json.loads(line)
    if '-9999' in line:
        continue
    else:
        if info["RFA_2F"] in RFA_2F:
            RFA_2F[info["RFA_2F"]] +=1
        else:
            RFA_2F[info["RFA_2F"]] =1
        if info["RFA_2A"] in RFA_2A:
            RFA_2A[info["RFA_2A"]] +=1
        else:
            RFA_2A[info["RFA_2A"]] =1
        if info["WEALTH_INDEX"] in WEALTH_INDEX:
            continue
        else:
            WEALTH_INDEX.append(info["WEALTH_INDEX"])
            
print("RFA_2F : ")
print(RFA_2F)
print("RFA_2A : ")
print(RFA_2A)


        
```

#### Part E)  - TARGET_B describes whether this customer responded (1) or did not respond (0) to the campaign.  What proportion of the targeted consumers responded?

#### Part F) - Is there a difference in the distribution of the WEALTH_INDEX for the responders from the non-responders?  Explain.

```python
import itertools
data = open("assess1_data.json", "r")

WI_Resp = dict()
WI_NonResp = dict()

for line in data:
    info = json.loads(line)
    if '-9999' in line:
        continue
    else:
        if info["TARGET_B"] == 1:
            if info["WEALTH_INDEX"] in WI_Resp:
                WI_Resp[info["WEALTH_INDEX"]] +=1
            else:
                WI_Resp[info["WEALTH_INDEX"]] =1
        elif info["TARGET_B"] ==0:
            if info["WEALTH_INDEX"] in WI_NonResp:
                WI_NonResp[info["WEALTH_INDEX"]] +=1
            else:
                WI_NonResp[info["WEALTH_INDEX"]] =1

print("Responders and Non-Reponders Distribution : ")
print("")
print("(Responders) (Non-Responders)")
for R,N in itertools.izip_longest(sorted(WI_Resp, key=WI_Resp.get, reverse=True),sorted(WI_NonResp, key=WI_NonResp.get, reverse=True),fillvalue='-'):
    if(R=='-'):
        print("(--,--)"),
    else:
        print(R,WI_Resp[R]),
    print("\t"),
    print(N,WI_NonResp[N])

```

#### Part J) - Data such as this are used to evaluate predictive models. Consider predicting response with the following quantity: -40 + 0.8*WEALTH_INDEX + 0.12*RFA_2F Print the top-10 customer records, ordered by this quantity.  Include the value of this quantity.

```python
data = open("assess1_data.json", "r")

Users = {}


for line in data:
    info = json.loads(line)
    if '-9999' in line:
        continue
    else:
        quant = float((-40)+(0.8*info["WEALTH_INDEX"])+(0.12*info["RFA_2F"]))
        #print(quant)
        Users[quant] = info

SortedUsers = sorted(Users.items(),key=lambda t: t[0], reverse=True)

count = 1
for u in SortedUsers:
    if count <= 10:
        print(str(count)+" : "),
        print(u[0]),
        print(","),
        print(u[1]["NAME"]),
        print("\n")
        count+=1
```
